```python
import pandas as pd
```

```python
df = pd.read_csv('entire_data.csv')
```

```python
df.head()
```

```python
ratings =  df.pivot_table(index=['userId'],columns=['title'],values='rating')
userRatings = ratings.fillna(0,axis=1)
corrMatrix = userRatings.corr(method='pearson')
```

```python
corrMatrix
```

```python
print(type(corrMatrix))
```

```python
corrMatrix.to_csv('corr_matrix.csv')
```
